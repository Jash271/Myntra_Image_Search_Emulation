```python
def Add_Image(path):
    df = pd.read_csv('data.csv')
    img=Image.open(path)
    new_img=img.resize((64,64),Image.ANTIALIAS)
    new_img.save('image_test.jpg')
    k=cv2.imread(f'image_test.jpg')
    k=k.reshape(-1,12288)
    print(k)
    
    return k
```
